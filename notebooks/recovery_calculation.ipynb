# Recovery Calculation

You need to simulate the process of gold recovery from gold ore. <br>
Use the following formula to simulate the recovery process:

Recovery = (C * (F - T) / F * (C - T))

Where:
- C: 
    - For rougher concentrate recovery → the proportion of gold in the concentrate immediately after flotation, or
    - For final concentrate recovery → the proportion of gold after the purification process.
- F: 
    - For rougher concentrate recovery → the proportion of gold in the feed before flotation, or
    - For final concentrate recovery → the proportion of gold in the concentrate immediately after flotation.
- T: 
    - For rougher concentrate recovery → the proportion of gold in the rougher tailings immediately after flotation, or
    - For final concentrate recovery → the proportion of gold in the tailings after purification

To predict the recovery coefficient, it is necessary to determine the proportion of gold in both the concentrate and the tailings. Note that both the rougher concentrate and the final concentrate are important.

### Evaluation Metrics
The sMAPE (Symmetric Mean Absolute Percentage Error) metric is defined as:

sMAPE = (1 / N) × Σᵢ₌₁ⁿ [ |yᵢ − ŷᵢ| / ( (|yᵢ| + |ŷᵢ|) / 2 ) ]

Where: 
- yᵢ : target value for observation i in the dataset used to evaluate model performance
- ŷᵢ : predicted value for observation i (e.g., in the test set)
- N : number of observations in the sample
- Σᵢ₌₁ⁿ : summation over all observations (with i ranging from 1 to N)

We need to predict two values:
1.	Rougher concentrate recovery: rougher rougher.output.recovery.
2.	Final concentrate recovery: final.output.recovery.

The final evaluation metric combines both values.

- sMAPE final : 25% * sMAPE(rougher) + 75% * sMAPE(final)
