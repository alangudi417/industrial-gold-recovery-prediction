# Technological Process

The extracted ore undergoes a primary treatment to obtain the ore mixture, also known as the rougher feed, which is the raw material used for flotation (also referred to as the rougher process).
After flotation, the material is subjected to a two-stage purification process.

1. Flotation
- The gold ore mixture is fed into flotation plants to obtain a rougher gold concentrate and rougher tailings (i.e., residues with a low concentration of valuable metals).
- The stability of this process is affected by the volatility and the unfavorable physico-chemical state of the flotation pulp (a mixture of solid particles and liquid).

2. Purification
- The rougher concentrate undergoes two purification stages. After this process, the final concentrate and new tailings are obtained.
